```python
%reload_ext autoreload
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

import statsmodels.api as sm
import scipy

# import own scripts
import sys
sys.path.insert(0, '/src/')

%autoreload 2
from utils import plot_multi, norm_dist_test, Model, clean_gdp, clean_ff, grid_search_SARIMA, plot_resid_SARIMA, one_ste_ahead_forecast, dynamic_forecast
```

```python
# load data
gdp = clean_gdp(pd.read_csv('data/GDPC1.csv'))
ff = clean_ff(pd.read_csv('data/FEDFUNDS.csv'))
```

## Grid Search for best parameters for ARIMA Model (no exogenous data)
---
> obtain optimal `order` and `seasonal_order` parameters

```python
model1 = grid_search_SARIMA(gdp)
```

```python
model1 = model1.fit(gdp)
model1.summary()
```

```python
# plot residual errors
plot_resid_SARIMA(model1, gdp)
```

## Rebuild ARIMA and Incorporate `FedFunds`
---
> integrate optimal `order` and `seasonal_order` parameters <br>
> integrate exog array with `FedFund`

```python
merged = gdp.join(ff).dropna(axis=0)
```

```python
model2 = sm.tsa.statespace.SARIMAX(merged['GDP'],
                                exog=merged['FedFundsRate'],
                                order=(1, 1, 2),
                                seasonal_order=(1, 1, 2, 12),
                                enforce_stationarity=False,
                                enforce_invertibility=False)

model2 = model2.fit()
```

```python
model2.summary()
```

```python
plot_resid_SARIMA(model2, gdp)
```

```python
one_ste_ahead_forecast(model2, gdp,'2000','2015-01-01')
```

```python
dynamic_forecast(model2, gdp, '2010', '2015-01-01')
```

```python
# Change in AIC
[model1.aic(), model2.aic]
```

```python
# Use KS test to see if there is a difference
scipy.stats.ks_2samp(model1.resid(), model2.resid)
```

# Interpretation of results:
> AIC dropped by ~20% <br>
> KS test reveals a p-value of .09 in the distribution of Residual Errors, (threshold is 0.05) <br>
> Therefore, `model2` which incorporated the `FedFunds` rate did indeed out perform `model1`, which had no exogenous features
